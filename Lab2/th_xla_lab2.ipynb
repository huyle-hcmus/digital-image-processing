### Sobel

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("data/coca.jpg")
img = img.convert("L")
img = np.array(img)

img_final = np.zeros((len(img),len(img[0])))
gx = np.zeros((len(img),len(img[0])))
gy = np.zeros((len(img),len(img[0])))

for i in range(1, len(img)-1): # Height
    for j in range(1, len(img[0]) - 1): # Width
        gx[i][j] = (img[i - 1][j - 1] + 2*img[i][j - 1] + img[i + 1][j - 1]) - (img[i - 1][j + 1] + 2*img[i][j + 1] + img[i + 1][j + 1])
        gy[i][j] = (img[i - 1][j - 1] + 2*img[i - 1][j] + img[i - 1][j + 1]) - (img[i + 1][j - 1] + 2*img[i + 1][j] + img[i + 1][j + 1])
        img_final[i][j] = min(255, np.sqrt(gx[i][j]**2 + gy[i][j]**2))
        
plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.subplot(222)
plt.imshow(gx, cmap='gray')
plt.title("Vertical Sobel")
plt.subplot(223)
plt.imshow(gy, cmap='gray')
plt.title("Horizontal Sobel")
plt.subplot(224)
plt.imshow(img_final, cmap='gray')
plt.title("Sobel")
```

### Prewitt

```python

img = Image.open("data/coca.jpg")
img = img.convert("L")
img = np.array(img, dtype=np.int32)

img_final_prewitt = np.zeros((len(img), len(img[0])))
gx_prewitt = np.zeros((len(img), len(img[0])))
gy_prewitt = np.zeros((len(img), len(img[0])))

for i in range(1, len(img) - 1):  # Height
    for j in range(1, len(img[0]) - 1):  # Width
        gx_prewitt[i][j] = (img[i - 1][j - 1] + img[i][j - 1] + img[i + 1][j - 1]) - (img[i - 1][j + 1] + img[i][j + 1] + img[i + 1][j + 1])
        gy_prewitt[i][j] = (img[i - 1][j - 1] + img[i - 1][j] + img[i - 1][j + 1]) - (img[i + 1][j - 1] + img[i + 1][j] + img[i + 1][j + 1])
        img_final_prewitt[i][j] = min(255, np.sqrt(gx_prewitt[i][j] ** 2 + gy_prewitt[i][j] ** 2))

plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.subplot(222)
plt.imshow(gx_prewitt, cmap='gray')
plt.title("Vertical Prewitt")
plt.subplot(223)
plt.imshow(gy_prewitt, cmap='gray')
plt.title("Horizontal Prewitt")
plt.subplot(224)
plt.imshow(img_final_prewitt, cmap='gray')
plt.title("Prewitt")
```

### Robert

```python
img = Image.open("data/coca.jpg")
img = img.convert("L")
img = np.array(img, dtype=np.int32)  # Chuyển đổi sang kiểu dữ liệu lớn hơn

img_final_robert = np.zeros((len(img), len(img[0])))
gx_robert = np.zeros((len(img), len(img[0])))
gy_robert = np.zeros((len(img), len(img[0])))

for i in range(1, len(img)-1):
    for j in range(1, len(img[0]) - 1):
        gx_robert[i][j] = img[i - 1][j - 1] - img[i][j] 
        gy_robert[i][j] = img[i - 1][j] - img[i][j - 1]
        img_final_robert[i][j] = min(255, np.sqrt(gx_robert[i][j]**2 + gy_robert[i][j]**2))

plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.subplot(222)
plt.imshow(gx_robert, cmap='gray')
plt.title("Vertical Robert")
plt.subplot(223)
plt.imshow(gy_robert, cmap='gray')
plt.title("Horizontal Robert")
plt.subplot(224)
plt.imshow(img_final_robert, cmap='gray')
plt.title("Robert")
```
